```python
import pandas as pd

df = pd.read_csv(
    r"C:\Users\mukhe\OneDrive\Desktop\coding\ecommerce\data\features\customer_features.csv"
)

df.head()
```

```python
features = [
    "recency_days",
    "frequency",
    "monetary"
]

X = df[features]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=50
)

clusters = kmeans.fit_predict(X_scaled)
```

```python
kmeans.n_features_in_
```

```python
import joblib

joblib.dump(
    kmeans,
    r"C:\Users\mukhe\OneDrive\Desktop\coding\ecommerce\models\kmeans_segmentation.joblib"
)

joblib.dump(
    scaler,
    r"C:\Users\mukhe\OneDrive\Desktop\coding\ecommerce\models\kmeans_scaler.joblib"
)
```
